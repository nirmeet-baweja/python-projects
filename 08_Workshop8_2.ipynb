#### <font color="red"><b>1.</b></font> Identify the base case in the mylen recursive function below then convert the function into an iterative version.
<code>
def mylen(p, n):
    if p == []: return n
    else: return mylen(p[1:], n+1)
</code>

```python
def myleni(p):
    temp_p = p[:]
    result = 0
    while temp_p != []:
        result = result+1
        temp_p = temp_p[1:]
    return result
```
